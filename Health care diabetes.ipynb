Data wrangling techniques

```python
# As shown in the Glucose Histogram most of data resides on the middle of the dataset we can fill missing values with mean
data['Glucose'] = data['Glucose'].replace(0,data['Glucose'].mean())

# As shown in the Glucose Histogram most of data resides on the middle of the dataset we can fill missing values with mean
data['BloodPressure'] = data['BloodPressure'].replace(0,data['BloodPressure'].mean())

# As shown in the Glucose Histogram most of data resides on the middle of the dataset we can fill missing values with mean
data['SkinThickness'] = data['SkinThickness'].replace(0,data['SkinThickness'].mean())

# As shown in the Glucose Histogram most of data resides on the middle of the dataset we can fill missing values with mean
data['BMI'] = data['BMI'].replace(0,data['BMI'].mean());

# because the data is skewed median will be best fit for replacing missing data 
data['Insulin'] = data['Insulin'].replace(0,data['Insulin'].median())
```

# Data Modeling

```python
X_features = data.drop('Outcome',axis=1)
y_target = data['Outcome']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X_features,y_target,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
x_train =std.fit_transform(x_train)
x_test = std.fit_transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV


param = {'n_neighbors':np.arange(1,30),'weights':('uniform','distance')}

knn = KNeighborsClassifier(n_jobs=-1)

clf = GridSearchCV(knn,param)
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)

from sklearn.metrics import classification_report,auc,roc_curve
accuracy_score = classification_report(y_test,y_pred)
print(accuracy_score)
print("Best parameter was",clf.best_params_)

print("\n","ROC Curve")
clf_prob=clf.predict_proba(x_test)
clf_prob1=clf_prob[:,1]
fpr,tpr,thresh=roc_curve(y_test,clf_prob1)
roc_auc_knn=auc(fpr,tpr)
plt.figure(dpi=80)
plt.title("ROC Curve")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.plot(fpr,tpr,'b',label='AUC Score = %0.2f'%roc_auc_knn)
plt.plot(fpr,fpr,'r--',color='red')
plt.legend()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

params = {'criterion':('gini','entropy')}

rf = RandomForestClassifier(n_estimators=1000,n_jobs=-1,random_state=42)
clf = GridSearchCV(rf,params)

clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)

from sklearn.metrics import classification_report,auc,roc_curve
accuracy_score = classification_report(y_test,y_pred)
print(accuracy_score)
print(clf.best_params_)

print("\n","ROC Curve")
clf_prob=clf.predict_proba(x_test)
clf_prob1=clf_prob[:,1]
fpr,tpr,thresh=roc_curve(y_test,clf_prob1)
roc_auc_knn=auc(fpr,tpr)
plt.figure(dpi=80)
plt.title("ROC Curve")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.plot(fpr,tpr,'b',label='AUC Score = %0.2f'%roc_auc_knn)
plt.plot(fpr,fpr,'r--',color='red')
plt.legend()
```
